```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0,len(data)):
    data_review=re.sub('[^a-zA-Z]',' ',data['message'][i])
    data_review=data_review.lower()
    data_review=data_review.split()
    data_review=[ps.stem(word) for word in data_review if word not in stopwords.words('english')]
    data_review=' '.join(data_review)
    corpus.append(data_review)
```

```python

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
x=cv.fit_transform(corpus).toarray()

y=pd.get_dummies(data['label'])
y=y.iloc[:,0].values
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
# from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
spam_detector=RandomForestClassifier()
spam_detector.fit(X_train, y_train)
y_pred= spam_detector.predict(X_test)

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
acc
```

```python
from sklearn import tree

spam_detector=tree.DecisionTreeClassifier()
spam_detector.fit(X_train, y_train)
y_pred= spam_detector.predict(X_test)

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
acc
```

```python
from sklearn.neighbors  import KNeighborsClassifier
spam_detector=KNeighborsClassifier()
spam_detector.fit(X_train, y_train)
y_pred= spam_detector.predict(X_test)

from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
acc
```
